```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def chebyshev(n, x):
    '''Función que calcula los polinomios de Chebyshev de orden n para valores x.'''
    if n == 0:
        y = np.ones_like(x)
    elif n == 1:
        y = x
    else:
        y = 2 * x * chebyshev(n - 1, x) - chebyshev(n - 2, x)
    return y
```

```python
def chebyshev(n, x):
    '''Función que calcula los polinomios de Chebyshev de orden n para valores x.'''
    m = len(x)
    T = np.zeros(m)
    T1 = np.ones(m)
    T2 = x.copy()

    if n == 1:
        T = T2
    else:
        for j in range(2, n + 1):
            T = 2 * x * T2 - T1
            T1, T2 = T2, T

    return T
```

```python
plt.figure()
plt.title('Polinomios de Chebyshev')

# Trazar los polinomios de Chebyshev
for n in range(1, 6):
    x_values = np.linspace(-1, 1, 100)
    y_values = chebyshev(n, x_values)
    plt.plot(x_values, y_values, label=f'Orden {n}')

plt.legend(loc='lower right')  # queremos colocarlo a la derecha de la gráfica
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
